<h2>Introduction/Bussiness problem</h2>

The client wants to open an italian restaurant in a city and he wants to choose a good location for it. In order to do this, the venues near the possible neighborhood locations must be studied taking into account the different kind of venues in the area (for example, the presence of other restaurants would be negative since there would be more competition in the area, while other venues like museums or parks would be beneficial since there would be more potential clients around the area).

Clustering the neighborhoods as it was done in the previous assignment would make the task easier since it would be faster to recognize which kind of venues are the most common in the neighorhoods of each cluster.

<h2>Data</h2>

The data we need would be similar to the data used in the previous assignment. It starts with the city that we want to study and its different neighborhoods with its coordinates, so the venues' data from the Foursquare API can be retrieved. Then the data would be treated in order to make the clusters based on the top 10 kind of venues for each area. After that, the venues can be studied and wheighted possitively or negatively based on the criteria that they would be beneficial or not to have them near the new restaurant, giving each neighborhood a score, so the highest score would be the better location.
